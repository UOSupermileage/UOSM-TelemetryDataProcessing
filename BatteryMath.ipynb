```python
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
from numpy import arange

# voltages =  [0, 1 , 2 , 3  , 4  , 5  , 6  , 7,   30  , 40  , 41  , 42  , 43  , 48  , 50  , 52  , 53,   54  , 56]
# codes =     [0, 45, 95, 145, 195, 245, 295, 345, 1485, 1980, 2030, 2080, 2130, 2370, 2470, 2565, 2615, 2660, 2755]

voltages =  [30  , 40  , 41  , 42  , 43  , 48  , 50  , 52  , 53,   54  , 56]
codes =     [1485, 1980, 2030, 2080, 2130, 2370, 2470, 2565, 2615, 2660, 2755]


def objective(x, a, b):
 return a * x + b

fig, ax = plt.subplots()
fig.set_size_inches(8,4)

popt, _ = curve_fit(objective, codes, voltages)
a, b = popt
print('y = %.5f * x + %.5f' % (a, b))
x_line = arange(min(codes), max(codes), 1)
y_line = objective(x_line, a, b)

plt.plot(x_line, y_line, '--', color='red')

ax.scatter(codes, voltages)
```
